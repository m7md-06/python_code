```python
# Titanic Data Exploration & Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Titanic dataset
titanic = pd.read_csv("Titanic.csv")
sns.set_style("whitegrid")

# Bar Plot
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x="Sex", hue="Survived", palette="Set2")
plt.title("Survival Count by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
#Histogram
plt.figure(figsize=(6, 4))
sns.histplot(data=titanic, x="Age", bins=30, kde=True, color="skyblue")
plt.title("Age Distribution of Passengers", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(data=titanic, x="Pclass", y="Age", palette="pastel")
plt.title("Age Distribution Across Passenger Classes", fontsize=14)
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python

# 4. Scatter Plot
plt.figure(figsize=(6, 4))
sns.scatterplot(data=titanic, x="Age", y="Fare", hue="Survived", style="Sex", palette="coolwarm")
plt.title("Fare vs Age by Survival Status", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()

```

```python
# Heatmap
plt.figure(figsize=(8, 6))
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```
